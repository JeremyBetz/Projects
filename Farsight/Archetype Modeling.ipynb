```python
import gensim
import pickle
```

```python
with open("Corpora.pkl", 'rb') as picklefile: 
    Class = pickle.load(picklefile)
```

```python
import itertools

Class = list(itertools.chain.from_iterable(Class))
```

```python
def game_reader():
    for game in range(len(Class)):
        hist = Class[game]
        yield (x for x in hist)
corpus = []
dictionary = gensim.corpora.Dictionary(game_reader())
for values in game_reader():
    corpus.append(dictionary.doc2bow(values))
```

```python
gensim.corpora.MmCorpus.serialize('Corpus.mm', corpus)
corpus = gensim.corpora.MmCorpus('Corpus.mm')
```

```python
archetypes=8
```

```python
lsi = gensim.models.LsiModel(corpus, id2word=dictionary, num_topics=archetypes)
lsi.print_topics()
```

```python
tfidf = gensim.models.TfidfModel(corpus, normalize=True)
corpus_tfidf = tfidf[corpus]
lsi_tfidf = gensim.models.LsiModel(corpus_tfidf, id2word=dictionary, num_topics=archetypes)
lsi_tfidf.print_topics()
```

```python
lda = gensim.models.LdaModel(corpus, id2word=dictionary, num_topics=archetypes)
lda.print_topics()
```
